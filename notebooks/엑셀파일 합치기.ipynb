```python
import pandas as pd
import glob

file_path = "/content/drive/MyDrive/지은/*.xlsx"
file_list = glob.glob(file_path)
df_list = []

for file in file_list:
    df = pd.read_excel(file, skiprows=1, nrows=10, header=None)

    df[1] = df[1].str.extract('5점만점에 (\d+)점')[0]
    df[1] = pd.to_numeric(df[1])

    df[3] = df[3].replace({
        "아주 만족해요": 3,
        "보통이에요": 2,
        "다소 아쉬워요": 1
    })

    df[4] = df[4].replace({
        "지속이 오래돼요": 3,
        "보통이에요": 2,
        "예상보다 짧아요": 1
    })

    df[5] = df[5].replace({
        "아주 만족해요": 3,
        "보통이에요": 2,
        "다소 아쉬워요": 1
    })

    df[6] = df[6].replace({
        "아주 촉촉해요": 3,
        "보통이에요": 2,
        "매트해요": 1
    })

    df_list.append(df)

final_df = pd.concat(df_list, ignore_index=True)

save_path = "/content/drive/MyDrive/지은/new!!!.xlsx"
final_df.to_excel(save_path, index=False, header=False)



print("모든 파일이 'new!!.xlsx'에 성공적으로 합쳐졌습니다.")
```
