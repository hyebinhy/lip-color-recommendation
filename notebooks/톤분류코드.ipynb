```python
import pandas as pd

# CSV 파일 경로
csv_file = '/content/drive/MyDrive/oceanpama.csv'

# CSV 파일에서 데이터 읽기
df = pd.read_csv(csv_file)
```

```python
df['tone'] = ""
```

```python
# dark_grayish
dark_grayish = df[(df.apply(lambda row: (row['saturation'] >= 0) and (row['saturation'] <= 24) and (row['lightness'] >= 15) and (row['lightness'] <= 34), axis=1))]
df.loc[dark_grayish.index, 'tone'] = 'dark_grayish'

# grayish
grayish = df[(df.apply(lambda row: (row['saturation'] >= 0) and (row['saturation'] <= 24) and (row['lightness'] >= 35) and (row['lightness'] <= 54), axis=1))]
df.loc[grayish.index, 'tone'] = 'grayish'

# light_grayish
light_grayish = df[(df.apply(lambda row: (row['saturation'] >= 0) and (row['saturation'] <= 24) and (row['lightness'] >= 55) and (row['lightness'] <= 74), axis=1))]
df.loc[light_grayish.index, 'tone'] = 'light_grayish'

# pale
pale = df[(df.apply(lambda row: (row['saturation'] >= 0) and (row['saturation'] <= 24) and (row['lightness'] >= 75) and (row['lightness'] <= 90), axis=1))]
df.loc[pale.index, 'tone'] = 'pale'

# dark
dark = df[(df.apply(lambda row: (row['saturation'] >= 25) and (row['saturation'] <= 49) and (row['lightness'] >= 15) and (row['lightness'] <= 34), axis=1))]
df.loc[dark.index, 'tone'] = 'dark'

# dull
dull = df[(df.apply(lambda row: (row['saturation'] >= 25) and (row['saturation'] <= 49) and (row['lightness'] >= 35) and (row['lightness'] <= 54), axis=1))]
df.loc[dull.index, 'tone'] = 'dull'

# soft
soft = df[(df.apply(lambda row: (row['saturation'] >= 25) and (row['saturation'] <= 49) and (row['lightness'] >= 55) and (row['lightness'] <= 74), axis=1))]
df.loc[soft.index, 'tone'] = 'soft'

# light
light = df[(df.apply(lambda row: (row['saturation'] >= 25) and (row['saturation'] <= 49) and (row['lightness'] >= 75) and (row['lightness'] <= 90), axis=1))]
df.loc[light.index, 'tone'] = 'light'

# deep
deep = df[(df.apply(lambda row: (row['saturation'] >= 50) and (row['saturation'] <= 74) and (row['lightness'] >= 15) and (row['lightness'] <= 44), axis=1))]
df.loc[deep.index, 'tone'] = 'deep'

# strong
strong = df[(df.apply(lambda row: (row['saturation'] >= 50) and (row['saturation'] <= 74) and (row['lightness'] >= 45) and (row['lightness'] <= 64), axis=1))]
df.loc[strong.index, 'tone'] = 'strong'

# bright
bright = df[(df.apply(lambda row: (row['saturation'] >= 50) and (row['saturation'] <= 74) and (row['lightness'] >= 65) and (row['lightness'] <= 90), axis=1))]
df.loc[bright.index, 'tone'] = 'bright'

# vivid
vivid = df[(df.apply(lambda row: (row['saturation'] >= 75) and (row['saturation'] <= 100) and (row['lightness'] >= 15) and (row['lightness'] <= 90), axis=1))]
df.loc[vivid.index, 'tone'] = 'vivid'
```

```python
import pandas as pd

# CSV 파일 경로
converted_colors_csv_path = '/content/drive/MyDrive/converted_colors.csv'
bright_csv_path = '/content/bright.csv'

# CSV 파일 불러오기
converted_colors_df = pd.read_csv(converted_colors_csv_path)
bright_df = pd.read_csv(bright_csv_path)

# 'File Name'을 인덱스로 설정하여 빠르게 검색할 수 있도록 변환
converted_colors_df.set_index('File Name', inplace=True)

# 'bright_df'에서 각 행을 순회하면서 'File Name'이 일치하는 경우 'color' 값을 업데이트
for idx, row in bright_df.iterrows():
    file_name = row['File Name']
    if file_name in converted_colors_df.index:
        bright_df.at[idx, 'color'] = converted_colors_df.at[file_name, 'color']

# 결과를 원래 파일에 저장
bright_df.to_csv('bright1.csv', index=False)
```
