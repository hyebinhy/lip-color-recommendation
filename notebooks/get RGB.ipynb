```python
import colorgram
import os

for file in files:
    img_path = os.path.join(folder_path, file)  # 이미지 파일 경로 생성
    colors = colorgram.extract(img_path, 3)  # 5개의 주요 색상을 추출

    print(f"{file}에서 추출된 색상들:")
    for color in colors:
               # color.proportion은 이미지에서 해당 색상이 차지하는 비율을 나타낸다.
        print(f"RGB: ({color.rgb.r}, {color.rgb.g}, {color.rgb.b}), 비율: {color.proportion:.2%}")

    print("")  # 파일 간 구분을 위해 빈 줄 출력
```

```python
import csv

with open('extracted_colors_max_proportion.csv', mode='w', newline='') as csv_file:
    writer = csv.writer(csv_file)
    writer.writerow(['File Name', 'Color Index', 'Red', 'Green', 'Blue', 'Max Proportion'])

    for file in files:
        img_path = os.path.join(folder_path, file)
        colors = colorgram.extract(img_path, 3)

        # 가장 높은 비율을 가진 색상 찾기
        max_color = max(colors, key=lambda c: c.proportion)

        # 가장 높은 비율을 가진 색상에 대한 정보를 CSV 파일에 쓰기
        writer.writerow([
            file, colors.index(max_color) + 1,  # 색상 인덱스 (1부터 시작)
            max_color.rgb.r, max_color.rgb.g, max_color.rgb.b,
            f"{max_color.proportion:.2%}"
        ])

print("색상 데이터가 'extracted_colors_max_proportion.csv' 파일에 저장.")
```

```python
import pandas as pd

# CSV 파일 경로
file_path = 'your_file.csv'  # 'your_file.csv'를 실제 파일 경로로 변경하세요.

# CSV 파일 읽기
df = pd.read_csv(file_path)

# RGB를 HEX 코드로 변환하는 함수, 여기서 'c', 'd', 'e'를 사용
def rgb_to_hex(row):
    return '#{:02x}{{:02x}{:02x}'.format(row['c'], row['d'], row['e'])

# RGB 값을 HEX 코드로 변환하고 결과를 새로운 컬럼에 저장
df['HEX'] = df.apply(rgb_to_hex, axis=1)

# 결과 확인
print(df)

# 변환된 데이터를 새로운 CSV 파일로 저장
df.to_csv('converted_colors.csv', index=False)
```

```python
import pandas as pd

def rgb_to_pccs(rgb):
    # PCCS 색상 코드와 해당 색상 범위 정의
    pccs_colors = {
        'PP': [(207, 114, 170), (250, 205, 221)],
        'RP': [(232, 74, 144), (248, 195, 216)],
        'RPB': [(189, 0, 97), (235, 166, 213)],
        'PB': [(67, 20, 100), (182, 144, 202)],
        'PBG': [(0, 119, 172), (179, 223, 232)],
        'BG': [(0, 134, 149), (171, 218, 215)],
        'B': [(0, 95, 125), (179, 215, 215)],
        'BGY': [(0, 151, 94), (175, 224, 170)],
        'YG': [(150, 195, 0), (225, 237, 132)],
        'GY': [(142, 151, 0), (217, 217, 97)],
        'YG': [(150, 195, 0), (225, 237, 132)],
        'Y': [(229, 225, 37), (249, 244, 139)],
        'YR': [(231, 158, 0), (249, 217, 123)],
        'R': [(228, 0, 15), (245, 144, 126)],
        'RP': [(232, 74, 144), (248, 195, 216)],
        'RPB': [(189, 0, 97), (235, 166, 213)]
    }

    # RGB값 추출
    r, g, b = rgb

    # PCCS 색상 코드 찾기
    for color_code, (lower_bound, upper_bound) in pccs_colors.items():
        if lower_bound[0] <= r <= upper_bound[0] and lower_bound[1] <= g <= upper_bound[1] and lower_bound[2] <= b <= upper_bound[2]:
            return color_code

    return None


# CSV 파일에서 데이터 읽어오기
csv_file = "색상표.csv"  # CSV 파일 경로
df = pd.read_csv(csv_file)

# RGB 값을 가지고 있는 열을 읽어와서 PCCS로 변환한 후, 새로운 열 추가
df['PCCS'] = df.apply(lambda row: rgb_to_pccs((row['R'], row['G'], row['B'])), axis=1)

# 변환된 데이터프레임을 CSV 파일로 저장
output_file = "색상표_with_PCCS.csv"  # 저장할 CSV 파일 경로
df.to_csv(output_file, index=False)

print("PCCS로 변환된 데이터를 저장했습니다.")
```
