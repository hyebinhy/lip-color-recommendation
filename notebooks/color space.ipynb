```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import colorsys

def rgb_to_hsv(rgb):
    r, g, b = rgb[0] / 255.0, rgb[1] / 255.0, rgb[2] / 255.0
    h, s, v = colorsys.rgb_to_hsv(r, g, b)
    return [h, s, v]


def plot_hsv_cone_from_csv(csv_file_path):
    # CSV 파일에서 데이터 읽어오기
    df = pd.read_csv(csv_file_path)

    # RGB 값을 배열로 변환
    rgb_values = df[['Red', 'Green', 'Blue']].values

    # RGB를 HSV로 변환
    hsv_values = [rgb_to_hsv(rgb) for rgb in rgb_values]
    hsv_values = pd.DataFrame(hsv_values, columns=['Hue', 'Saturation', 'Value'])

    # 원뿔 모양의 HSV 공간 시각화
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    # HSV 값을 원뿔 모양으로 변환
    x = hsv_values['Saturation'] * hsv_values['Value'] * np.cos(hsv_values['Hue'] * 2 * np.pi)
    y = hsv_values['Saturation'] * hsv_values['Value'] * np.sin(hsv_values['Hue'] * 2 * np.pi)
    z = hsv_values['Value']

    # 산점도 그리기
    ax.scatter(x, y, z, c=rgb_values/255.0, marker='o')

    # 축 이름 설정
    ax.set_xlabel('Saturation * Value * cos(Hue)')
    ax.set_ylabel('Saturation * Value * sin(Hue)')
    ax.set_zlabel('Value')
    ax.set_title('HSV Color Space (Cone)')

    plt.show()

# CSV 파일 경로
csv_file_path = '/content/drive/MyDrive/converted_colors.csv'

# 원뿔 모양의 HSV 공간 시각화
plot_hsv_cone_from_csv(csv_file_path)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import colorsys

def rgb_to_hsv(rgb):
    r, g, b = rgb[0] / 255.0, rgb[1] / 255.0, rgb[2] / 255.0
    h, s, v = colorsys.rgb_to_hsv(r, g, b)
    return [h * 360, s, v]

def plot_hsv_sphere_from_csv(csv_file_path):
    # CSV 파일에서 데이터 읽어오기
    df = pd.read_csv(csv_file_path)

    # RGB 값을 배열로 변환
    rgb_values = df[['Red', 'Green', 'Blue']].values

    # RGB를 HSV로 변환
    hsv_values = [rgb_to_hsv(rgb) for rgb in rgb_values]
    hsv_values = pd.DataFrame(hsv_values, columns=['Hue', 'Saturation', 'Value'])

    # 3D 구로 HSV 공간 시각화
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    # HSV 값을 3D 구로 변환
    r = hsv_values['Value']
    theta = (1 - hsv_values['Saturation']) * np.pi
    phi = hsv_values['Hue'] / 180 * np.pi

    # 3D 산점도 그리기
    x = r * np.sin(theta) * np.cos(phi)
    y = r * np.sin(theta) * np.sin(phi)
    z = r * np.cos(theta)
    ax.scatter(x, y, z, c=rgb_values/255.0, marker='o')

    # 축 범위 설정
    ax.set_xlim([-1, 1])
    ax.set_ylim([-1, 1])
    ax.set_zlim([-1, 1])

    # 축 이름 설정
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_title('HSV Color Space (Sphere)')

    plt.show()

# CSV 파일 경로
csv_file_path = '/content/drive/MyDrive/converted_colors.csv'

# HSV 공간에 대한 구 형태의 시각화
plot_hsv_sphere_from_csv(csv_file_path)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import colorsys

def rgb_to_hsv(rgb):
    r, g, b = rgb[0] / 255.0, rgb[1] / 255.0, rgb[2] / 255.0
    h, s, v = colorsys.rgb_to_hsv(r, g, b)
    return [h * 360, s, v]

def plot_hsv_surface_from_csv(csv_file_path):
    # CSV 파일에서 데이터 읽어오기
    df = pd.read_csv(csv_file_path)

    # RGB 값을 배열로 변환
    rgb_values = df[['Red', 'Green', 'Blue']].values

    # RGB를 HSV로 변환
    hsv_values = [rgb_to_hsv(rgb) for rgb in rgb_values]
    hsv_values = pd.DataFrame(hsv_values, columns=['Hue', 'Saturation', 'Value'])

    # HSV 값을 3D 공간으로 변환
    r = hsv_values['Value']
    theta = (1 - hsv_values['Saturation']) * np.pi
    phi = hsv_values['Hue'] / 180 * np.pi

    # 3D 곡면 플로팅
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    x = r * np.sin(theta) * np.cos(phi)
    y = r * np.sin(theta) * np.sin(phi)
    z = r * np.cos(theta)

    # HSV 값을 사용하여 색상 지정
    ax.plot_trisurf(x, y, z, cmap='hsv', linewidth=0.2)

    # 축 범위 설정
    ax.set_xlim([-1, 1])
    ax.set_ylim([-1, 1])
    ax.set_zlim([-1, 1])

    # 축 이름 설정
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_title('HSV Color Space (Surface)')

    plt.show()

# CSV 파일 경로
csv_file_path = '/content/drive/MyDrive/converted_colors.csv'

# HSV 공간에 대한 곡면 플로팅 및 컬러 맵 할당
plot_hsv_surface_from_csv(csv_file_path)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import colorsys

def rgb_to_hsv(rgb):
    r, g, b = rgb[0] / 255.0, rgb[1] / 255.0, rgb[2] / 255.0
    h, s, v = colorsys.rgb_to_hsv(r, g, b)
    return [h * 360, s, v]

def plot_hsv_surface_from_csv(csv_file_path, min_rgb=[0, 0, 0], max_rgb=[255, 255, 255]):
    # CSV 파일에서 데이터 읽어오기
    df = pd.read_csv(csv_file_path)

    # RGB 값을 배열로 변환
    rgb_values = df[['Red', 'Green', 'Blue']].values

    # RGB 값을 범위 내에 있는 값만 필터링
    filtered_rgb_values = [rgb for rgb in rgb_values if min_rgb[0] <= rgb[0] <= max_rgb[0] and min_rgb[1] <= rgb[1] <= max_rgb[1] and min_rgb[2] <= rgb[2] <= max_rgb[2]]

    # RGB를 HSV로 변환
    hsv_values = [rgb_to_hsv(rgb) for rgb in filtered_rgb_values]
    hsv_values = pd.DataFrame(hsv_values, columns=['Hue', 'Saturation', 'Value'])

    # HSV 값을 3D 공간으로 변환
    r = hsv_values['Value']
    theta = (1 - hsv_values['Saturation']) * np.pi
    phi = hsv_values['Hue'] / 180 * np.pi

    # 3D 곡면 플로팅
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    x = r * np.sin(theta) * np.cos(phi)
    y = r * np.sin(theta) * np.sin(phi)
    z = r * np.cos(theta)

    # HSV 값을 사용하여 색상 지정
    ax.plot_trisurf(x, y, z, cmap='hsv', linewidth=0.2)

    # 축 범위 설정
    ax.set_xlim([-1, 1])
    ax.set_ylim([-1, 1])
    ax.set_zlim([-1, 1])

    # 축 이름 설정
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_title('HSV Color Space (Surface)')

    plt.show()

# CSV 파일 경로
csv_file_path = '/content/drive/MyDrive/converted_colors.csv'


# 원하는 RGB 값의 범위 설정
min_rgb = [0, 0, 0]  # 최소 RGB 값
max_rgb = [255, 255, 255]  # 최대 RGB 값

# HSV 공간에 대한 곡면 플로팅 및 컬러 맵 할당
plot_hsv_surface_from_csv(csv_file_path, min_rgb, max_rgb)
```
