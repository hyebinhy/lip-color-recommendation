```python
import pandas as pd

csv_file = '/content/drive/MyDrive/전처리데이터/review.csv'
df = pd.read_csv(csv_file, encoding='utf-8-sig')

df.tail()
```

```python
df1 = df[['name_product', 'hydration']]
df1.tail()
```

```python

# 'name_product'별로 'longevity'의 평균 계산
df_avg = df.groupby('name_product')['hydration'].mean().reset_index()

# 결과를 새로운 파일로 저장
output_file = '/content/drive/MyDrive/전처리데이터/average_hydration.csv'
df_avg.to_csv(output_file, index=False, encoding='utf-8-sig')

# 결과 확인
df_avg.tail()
```
